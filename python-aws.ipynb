```python
pelicula = [
    "The Shawshank Redemption",
    "The Dark Knight",
    "Inception",
    "Pulp Fiction",
    "The Matrix",
    "Interstellar",
    "Parasite",
    "Gladiator",
    "Titanic",
    "The Lord of the Rings: The Fellowship of the Ring"
]

estreno = [
    1994,
    2008,
    2010,
    1994,
    1999,
    2014,
    2019,
    2000,
    1997,
    2001
]

vista = [
    True,
    False,
    True,
    False,
    True,
    False,
    True,
    False,
    True,
    False
]

def imprimir_peliculas_numeradas(peliculas, estrenos, vistas):
    for i, (nombre, anio, se_vio) in enumerate(zip(peliculas, estrenos, vistas), start=1):
        estado = "VISTA" if se_vio else "NO VISTA"
        print(f"{i}. {nombre} ({anio}) - {estado}")

imprimir_peliculas_numeradas(pelicula, estreno, vista)
```
